4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.

```python
import pandas as pd
```

```python
df_orders = pd.read_csv("orders.csv")
df_orders
```

```python
#1
df_orders['order_date'] = pd.to_datetime(df_orders['order_date'], format="%Y-%m-%d")
df_orders['order_date'].sample(1)
```

```python
#2
print("Unique values:")
print(df_orders['tshirt_category'].unique())
print("Number of unique values:")
print(df_orders['tshirt_category'].nunique())
```

```python
#3
df_orders['tshirt_category'] = df_orders['tshirt_category'].apply(
    lambda item : item\
        .replace("Tshirt", "T-Shirt")\
        .replace("Wh T-Shirt M", "White T-Shirt M")\
        .replace("Bl T-Shirt M", "Black T-Shirt M")\
        .replace("Wh T-Shirt F", "White T-Shirt F")\
        .replace("Bl T-Shirt F", "Black T-Shirt F")\
        .replace("Tennis Shirt", "Tennis_Shirt")
)
df_orders['tshirt_category'].value_counts()
```

```python
df_orders['tshirt_category'] = df_orders['tshirt_category'].apply(lambda x: x if (x in ['Hoodie', 'Tennis_Shirt']) else f'{x.split(" ")[1]} {x.split(" ")[0]} {x.split(" ")[2]}')
df_orders[['tshirt_type', 'tshirt_colour', 'tshirt_size']] = df_orders['tshirt_category'].str.split(expand=True)
df_orders
```

```python
#5
df_orders[(df_orders['order_date'] >= '2014-12-31') & (df_orders['order_date'] <= '2016-08-02')].sort_values(by='order_date')
```
